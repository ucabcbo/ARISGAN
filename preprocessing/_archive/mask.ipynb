```python
import geopandas as gpd
from shapely.geometry import Polygon

gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'

# Replace 'path/to/your/file.kml' with the actual path to your KML file
kml_file = '/home/cb/sis2/data/17NXA_SeaIceMask.kml'

# Read the KML file using geopandas
gdf = gpd.read_file(kml_file)

# Assuming the KML file contains only one Polygon, you can access its geometry like this
polygon_geometry = gdf.geometry[0]

# Convert the polygon geometry to a Shapely Polygon
shapely_polygon = Polygon(polygon_geometry.exterior.coords)

print(shapely_polygon)
```

```python
from matplotlib import pyplot as plt

# Plot the Shapely Polygon using matplotlib
fig, ax = plt.subplots(figsize=(8, 8))
x, y = shapely_polygon.exterior.xy
ax.plot(x, y, color='blue', linewidth=2)
ax.set_aspect('equal')
plt.show()
```
